## Ejercicio 1: Gráfico de líneas con múltiples líneas y leyenda

Crear gráfico de líneas con múltiples líneas y leyenda

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 6, 9, 12, 15]

fig = plt.figure(figsize=(10,6))
plt.plot(x, y1, label='Line 1', color='r')
plt.plot(x, y2, label='Line 2')

plt.title('Gráfico de líneas con múltiples')
plt.xlabel('X')
plt.ylabel('Y')

plt.legend()
plt.show()
```

## Ejercicio 2: Gráfico de dispersión con tamaño y color variables
Crear gráfico de dispersión con tamaño y color variables

```python
x = np.random.rand(100)
y = np.random.rand(100)

tamanio = np.random.rand(100) * 300
color = np.random.rand(100)

fig = plt.figure(figsize=(10,5))


plt.scatter(x, y, s=tamanio, c=color)


plt.title('Gráfico de dispersión con tamaño y color variables')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar()
plt.show()
```

## Ejercicio 4: Gráfico de torta con porcentajes y separación
Crear gráfico de torta con porcentajes y separación

```python
porcentajes = [30, 20, 25, 15, 10]
etiquetas = ['A', 'B', 'C', 'D', 'E']
separacion = [0.1, 0, 0, 0, 0] 

fig, ax = plt.subplots()
ax.pie(porcentajes, labels=etiquetas, autopct='%1.1f%%', explode=separacion)

ax.set_title('Crear gráfico de torta con porcentajes y separación')
plt.axis('Equal')
plt.show()


```

## Ejercicio 5: Gráfico de barras apiladas con colores personalizados
Crear gráfico de barras apiladas con colores personalizados

Mostrar el gráfico

```python
# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores1 = [15, 7, 10, 12]
valores2 = [8, 10, 6, 9]

fig = plt.figure(figsize=(10,5))

# Colores personalizados
colores = ['r', 'g', 'b', 'y']


plt.bar(categorias, valores1, color = colores, label = 'Valores 1')
plt.bar(categorias, valores2,  alpha= 0.5, label = 'Valores 2')

plt.title('gráfico de barras apiladas con colores personalizados')
plt.xlabel('categorias')
plt.ylabel('Valores')
plt.legend(['valores1', 'valores2'])
plt.show()



```

```python
titanic = pd.read_excel('./titanic.xls')
```

```python
df = titanic
df.sample(10)
```

```python
sobrevivientes = df['survived'].value_counts()
tripulacion = ['fallecidos', 'sobrevivientes']

fig = plt.figure(figsize=(10,5))

fig, ax = plt.subplot()
ax.pie(sobrevivientes, labels=tripulacion, autopct='%1.1f%%', explode=[0.2, 0])

ax.set_title('Distribución de Sobrevivientes y Fallecidos')

plt.show()
```
