#### Cuál es el jugador que más tiempo ha jugado y cuál es la media de tiempo de todos los jugadores?  
Esta pregunta tiene truco porque hay muchos jugadores suplentes que no juegan casi nunca y van a sesgar los datos, cómo podrías solucionar esto?


```python
max_time_index = df['time'].idxmax()
max_time_played = df.loc[max_time_index, 'player_name']

players_mean = str(df['time'].mean())

print('el jugador que más tiempo ha jugado es: ' + max_time_played)
print('Media del tiempo jugado de todos los jugadores: ' + players_mean)
```

#### Qué posición es la que más tiempo juega? Y la que menos?


```python
time_played_position = df.groupby('position')['time'].sum()
position_max_time_played = str(time_played_position.idxmax())


position_min_time_played = str(time_played_position.idxmin())


print(time_played_position)
print('\nLa posición que mas tiempo juega es: ' + position_max_time_played)
print('La posición que menos tiempo juega es: ' + position_min_time_played)
```

#### Qué porcentaje de goles son hechos por penales? La columna npg significa: goles hechos SIN penales.


```python
total = df['goals'].sum()
npg = df['npg'].sum()
pg = total - npg

persentage = str((pg * 100) / total)

print('Un ' + persentage + '% de goles son por penales.')
```

#### Qué porcentaje de goles son hechos con asistencias?


```python
total_goals = df['goals'].sum()
assists = df['assists'].sum()

persentage = str((assists * 100) / total_goals)


print('Un ' + persentage + '% de goles son por asistencias.')
```

#### Cuál fue el equipo que más goles hizo cada año? Salió campeón ese año?


```python

team_goals = df.groupby(['year', 'team_name'])['goals'].sum()
team_goals = team_goals.reset_index()

max_goals = team_goals.groupby('year')['goals'].idxmax()

team_max_goals = team_goals.loc[max_goals]
team_max_goals = team_max_goals.sort_values('year', ascending=False)

for year, team_name, goals in team_max_goals[['year', 'team_name', 'goals']].values:
    print(f"El equipo que más goles hizo en {year} fue {team_name} con {goals} goles.")
    
```

#### Cuántos jugadores jugaron todos los años del dataset? Qué porcentaje del total representan?

```python
years_played = df.groupby('player_name')['year'].nunique()

every_year = years_played[years_played == df['year'].nunique()]

#years_played
#every_year

total_every_year = len(every_year)

total = len(years_played)


persentage = (total_every_year * 100) / total


print(f'{total_every_year} jugadores jugaron todos los años.')
print(f'Los jugadores que jugaron todos los años representasn el {persentage}%.')
```
