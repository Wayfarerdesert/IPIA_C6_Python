```python
query_create_table = """
CREATE TABLE IF NOT EXISTS usuarios (
  id int(11) NOT NULL AUTO_INCREMENT,
  Titulo varchar(50),
  Cargo varchar(50),
  PRIMARY KEY (id)
);

"""
```

```python
# Cramos el nombre de las columnas
name_colum = "`,`".join([str(i) for i in df.columns.tolist()])

# iteramos en cada linea del df para tomar cada uno de los valores e ingestarlos
inicio = time.time()
for i, row in df.iterrows():
    sql = "INSERT INTO `usuarios` (`" +name_colum + "`) VALUES (" + "%s,"*(len(row)-1) + "%s)"
    cursor.execute(sql, tuple(row))
    connection.commit()
fin = time.time()
print(fin-inicio)
```
